----- Import all library functions-----

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

---- Read the Dataset-----

```python
data=pd.read_csv('imdb_top_1000.csv')
```

----- display the Dataset ----

```python
data
```

----- Display the information of the raw dataset ----

```python
data.info()
```

----- Remove the unnecessary columns -----

```python
data.drop(['Poster_Link','Overview'],axis=1,inplace=True)
```

----- Cleaning of the data-----

```python
#To make the Released year as int type 
data['Released_Year'].unique()
```

```python
data= data[data['Released_Year']!='PG']
data['Released_Year']=data['Released_Year'].astype(int)
data['Runtime']=data['Runtime'].replace('min','',regex=True).astype(int)
data['Gross']=data['Gross'].replace(',','', regex=True).fillna(0).astype(float)
data.dropna(inplace=True)
```

------Average Earning of Top 20 Directors ------

```python

pd.options.display.float_format='{:2f}'.format
avg=data.groupby('Director')['Gross'].mean().sort_values(ascending=False).head(20)
Dir_Avg_gross=avg.map(lambda x:'%.2f'%x)
Dir_Avg_gross
```

```python
x=data.groupby('Director')['Gross'].mean().sort_values(ascending=False).head(20)
mylabels=x.index.tolist()
plt.figure(figsize=(15,10))
color=plt.cm.tab20(np.linspace(0,1,len(mylabels)))
for a,c,x in zip(mylabels,color,x):
    plt.bar(a,x,width=0.5,color=c, edgecolor='g',linewidth=0,label=a)
plt.xlabel('Director',fontsize=20)
plt.ylabel('Income in core', fontsize=25)
plt.title('Directors Average Income',fontsize=20)
plt.legend(loc='upper right')
plt.xticks([])
plt.show()
```

------ Number of Movies Released every Year -----

```python
data.groupby('Released_Year')['Series_Title'].count().sort_values(ascending=False)
```

```python
y=data.groupby('Released_Year')['Series_Title'].count().sort_values(ascending=False).head(20)
x=y.index.tolist()
fig=plt.figure(figsize=(20,15))
colors=['purple' if (bar==max(y)) else 'skyblue' for bar in y]
plt.bar(x,y,color=colors,label=y)
plt.title('Movie count based on year',fontsize=20)
plt.ylabel('Number of movies',fontsize=20)
plt.xlabel('Released Year',fontsize=20)
plt.xticks(x)
plt.show()
```

-----Top 10 Average Rating of Directors -----

```python
avg=data.groupby('Director')['IMDB_Rating'].mean().sort_values(ascending=False).head(10)
total_ratings=avg.map(lambda x: '%.2f' % x)
total_ratings
```

```python
total_votes=data.groupby('Director')['No_of_Votes'].sum().sort_values(ascending=False).head(10)
tot=data['No_of_Votes'].sum()
per_vote=[(total_votes/tot)*100]
per_vote
```

------ Different Genre of the the movies------

----- Different Certificates of the movies -----

```python
x=data.groupby('Certificate')['Certificate'].count().sort_values(ascending=False)
mylabels=x.index.tolist()
plt.pie(x,labels=mylabels)
plt.show()
```

----- Yearly Revenue from the Movies-----

```python
g=data.groupby('Released_Year')['Gross'].sum().sort_values(ascending=False).head(20)
mydata=g.index.tolist()
plt.figure(figsize=(15,12))
plt.bar(mydata,g, color='lightpink')
plt.xticks(mydata)
plt.yticks()
plt.xlabel('Released Year',fontsize=15)
plt.ylabel('Revenue of the Year',fontsize=15)
plt.title('Yearly Revenue', fontsize=15)
plt.show()
```
